## Question #1

(a) In a matrix $x$, generate 500 samples, each of size $n = 50$, from your favorite (non-normal) distribution.

#### (Note! In any practical scenario, we would only have one sample of size $n$. The purpose here is the study the distribution of the sample mean over many (here, 500) samples. This is an extremely important concept. The distribution of the sample mean ("xbar"), as well as the distribution of other estimators, over many samples is heavily used in the statistical inference portion of the class)

```r
n = 50; m = 500; p = 0.5;
x = replicate(m, expr = rbinom(n,n,p))
dim(x)
```

(b) Calculate the mean of each sample. This should be a vector of size 500. Call this vector xbar.

#### xbar represents having the mean of 500 different samples (again: in the real world, we'd only have one of these samples; but this allows us to think about the distribution of the estimator "xbar").


```r
xbar = colMeans(x)
length(xbar)
```

(c) Construct a histogram of the means. What do you notice? What is the name of this distribution, i.e., the distribution of "xbar"? What is the mean of "xbar"? The variance of "xbar"? 


```r
x = seq(23,27, length.out = 1000)
hist(xbar, freq = FALSE, ylim = c(0,0.9))
lines(x,dnorm(x, mean = n*p, sd = sqrt(p*(1-p))))

mean(xbar)
var(xbar)

theorymean = n*p; print(theorymean)
theoryvar = n*p*(1-p)/n; print(theoryvar)
```

## Question #2

About two out of every three purchases at Monk's Cafe are paid for by credit cards. Four hundred and eighty customers buying food at Monk's are randomly selected. 

(a) Find $n$, $p$, ($p$ = probability that someone pays by credit card), and $q = 1-p$.



```r
n = 480; p = 2/3; q = 1-p;
```

(b) Find the probability that greater than 300 will pay for their purchases using credit card using a binom function.

```r
1 - pbinom(300, n, p) #1-[F(300)]
```

(c) Now, consider X ~ N(np, npq). Find the probability that X is greater than 300. Compare this to the answer to part (b).


```r
1 - pnorm(300, mean = n*p, sd = sqrt(n*p*q))
```

(d) Plot the pmf of the binomial random variable used in (a) and (b) and the pdf of X from part (c). Comment on the plots. 


```r
x = seq(250, 400); yb = dbinom(x, n, p); yn = dnorm(x, n*p, sqrt(n*p*q))

par(mfrow = c(2,1))
plot(x,yb, type = "l", xlab = "Binomial Probabilities")
plot(x,yn, type = "l", xlab = "Normal Probabilities")
```

They look the same!
